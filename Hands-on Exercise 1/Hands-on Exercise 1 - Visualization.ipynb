## Style Calls

One of the main reasons to learn the absolute basics is to see how the style interactions effect the API.

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

```python
plt.plot(x,y)

# Axis and ticks
plt.xlim(0,2)
plt.ylim(100,300)


# Labeling
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

```python
plt.plot(x,y,color='red')

# Axis and ticks
plt.xlim(0,2)
plt.ylim(100,300)


# Labeling
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

```python
plt.plot(x,y,color='red',marker='o',markersize=20,linestyle='--')

# Axis and ticks
plt.xlim(0,2)
plt.ylim(100,300)


# Labeling
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

## The Data

Context

The "goal" field refers to the presence of heart disease in the patient. It is integer valued from 0 (no presence) to 4.

Columns
* age  age in years
* sex  (1 = male; 0 = female)
* cp  chest pain type
* trestbps   resting blood pressure (in mm Hg on admission to the hospital)
* cholserum   cholestoral in mg/dl
* fbs(fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
* restecg   resting electrocardiographic results
* thalach   maximum heart rate achieved
* exang    exercise induced angina (1 = yes; 0 = no)
* oldpeak   ST depression induced by exercise relative to rest
* slope    the slope of the peak exercise ST segment
* ca    number of major vessels (0-3) colored by flourosopy
* thal3 = normal; 6 = fixed defect; 7 = reversable defect
* target   1 or 0

## Pairplots

INFO: https://seaborn.pydata.org/generated/seaborn.pairplot.html
