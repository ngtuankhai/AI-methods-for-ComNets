```python
G = nx.Graph()
G.add_node(1)
# add nodes from any iterable container, such as a list
G.add_nodes_from([2,3])
```

Nodes from one graph can be incorporated into another.

```python
T = nx.path_graph(10)
G.add_nodes_from(T)
nx.draw_networkx(G)
```

## Dominating Set

```python
from networkx.algorithms import dominating
G = nx.Graph()

G.add_nodes_from([1,2,3,4,5])

G.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 4), (4, 5)])

# Find the dominating set

dominating_set = nx.dominating_set(G)

plt.figure(figsize=(10,6))
pos = nx.spring_layout(G)

# Draw the nodes

nx.draw_networkx_nodes(G,pos,nodelist=dominating_set,node_color='red',node_size=500)
nx.draw_networkx_nodes(G,pos, nodelist = list(set(G.nodes()) - set(dominating_set)),node_color='lightblue',node_size=500)

# Draw the edges

nx.draw_networkx_edges(G,pos)

# Draw labels
nx.draw_networkx_labels(G,pos)

# Print the dominating set
print("Dominating Set: ",dominating_set)
```

## Metric Closure

```python
# Create a graph
G = nx.Graph()

# Add nodes to the graph
G.add_nodes_from(['A', 'B', 'C', 'D', 'E'])

# Add edges to the graph
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('C', 'D'), ('C', 'E'), ('D', 'E')])

# Compute the metric closure
closure = approx.metric_closure(G)

# Print the metric closure
print("Metric closure:")
for u in closure:
    for v in closure[u]:
        print(f"{u} -> {v}: {closure[u][v]['distance']}")

# Draw the original graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500)
plt.title('Original Graph')
plt.show()

# Draw the metric closure
M = nx.Graph()
for u in closure:
    for v in closure[u]:
        M.add_edge(u, v, weight=closure[u][v]['distance'])

pos_M = nx.spring_layout(M)
# Draw nodes
nx.draw_networkx_nodes(M, pos_M, node_color='lightblue', node_size=500)

# Draw edges
nx.draw_networkx_edges(M, pos_M)

# Draw labels
nx.draw_networkx_labels(M, pos_M)

# Draw edge weights
edge_labels = nx.get_edge_attributes(M, 'weight')
nx.draw_networkx_edge_labels(M, pos_M, edge_labels=edge_labels)

plt.title('Graph with Metric Closure and Edge Weights')
plt.show()
```

## Steiner Tree

```python
G = nx.Graph()

G.add_nodes_from(['A','B','C','D'])

G.add_weighted_edges_from([('A', 'B', 5.0), ('A', 'C', 2.0), ('B', 'D', 20.0), ('C', 'D', 10.0)])

# Define the terminal nodes for the Steiner Tree

terminal_nodes = ['A','D']

# Compute the Steiner Tree

steiner_tree = approx.steiner_tree(G,terminal_nodes)




# Draw the original graph
pos = nx.spring_layout(G)
nx.draw(G,pos,with_labels=True,node_color='lightblue',node_size=500)
nx.draw_networkx_edges(G,pos,width=1.5)
plt.title('Original Graph')
# Draw edge labels
edge_labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels = edge_labels)
plt.show()

# Draw the Steiner Tree

nx.draw(steiner_tree,pos,with_labels=True,node_color='lightblue',node_size=500)
nx.draw_networkx_edges(steiner_tree,pos,width=1.5)

# Draw edge labels
edge_labels = nx.get_edge_attributes(steiner_tree,'weight')
nx.draw_networkx_edge_labels(steiner_tree,pos,edge_labels = edge_labels)

plt.title('Steiner Tree')

plt.show()
```

## Vertex Cover

```python
# Create a graph
G = nx.Graph()

# Add nodes to the graph
G.add_nodes_from(['Alice', 'Bob', 'Charlie', 'Dave', 'Eve', 'Frank'])

# Add edges to the graph
G.add_edges_from([('Alice', 'Bob'), ('Alice', 'Charlie'), ('Charlie', 'Dave'), ('Charlie', 'Eve'), ('Dave', 'Eve'), ('Dave', 'Frank')])

# Assign weights to nodes
weights = {'Alice': 2, 'Bob': 1, 'Charlie': 3, 'Dave': 4, 'Eve': 2, 'Frank': 1}
nx.set_node_attributes(G,weights,'weight')

# Compute the minimum weighted vertex cover using min_weighted_vertex_cover()
vertex_cover = approx.min_weighted_vertex_cover(G,weight='weight')

# Print Minimum Vertex Cover
print("Minimum Vertex Cover:")
for node in vertex_cover:
  print(f"-{node}")
```

```python
from pulp import LpProblem, LpMinimize, LpVariable, lpSum

# Create a linear programming problem
problem = LpProblem("Minimum_Weighted_Vertex_Cover",LpMinimize)

# Create variables for each node
variables = LpVariable.dicts("x",G.nodes(),cat='Binary')

# Set the objective function
problem += lpSum(variables[node] * weights[node] for node in G.nodes())

# Add Constraints
for edge in G.edges():
  problem += variables[edge[0]] + variables[edge[1]] >= 1

# Solve the problem
problem.solve()

# Get the minimum weighted vertex cover nodes
vertex_cover = [node for node in G.nodes() if variables[node].value() == 1]

# Draw the original graph
plt.figure(figsize=(8,6))
pos = nx.spring_layout(G)
nx.draw(G,pos,with_labels=True,node_color='lightblue',node_size=500)

# Draw the vertex cover nodes in a different color
nx.draw_networkx_nodes(G,pos,nodelist=vertex_cover,node_color='salmon',node_size=500)

plt.title('Original Graph with Minimum Weighted Vertex Cover')
plt.show()

# Print the minimum weighted vertex cover
print("Minimum Weighted Vertex Cover:")
for node in vertex_cover:
    print(f"- {node}")

```

### Implementation of the Average Degree Connectivity

```python
from networkx.algorithms.assortativity.connectivity import average_degree_connectivity
# Create a graph
G = nx.Graph()
# Add nodes to the graph
G.add_nodes_from([1, 2, 3, 4, 5, 6, 7, 8])

# Add edges to the graph
G.add_edges_from([(1, 2), (1, 3), (2, 3), (2, 4), (3, 4), (4, 5),  (5, 7), (6, 7), (7, 8)])

nx.draw(G,with_labels=True)

# Compute the degree of each node
degrees = dict(G.degree())

# Compute the average degree connectivity for each node
avg_degree_connectivity = {}

for node in G.nodes:
  degree = degrees[node]
  neighbors = G.neighbors(node)
  neighbors_degrees = [degrees[neighbor] for neighbor in neighbors]
  avg_degree_connectivity[node] = sum(neighbors_degrees) / degree

# Calculate the average degree connectivity for each distinct degree value

degree_connectivity_dict = {}
for node,connectivity in avg_degree_connectivity.items():
  degree = degrees[node]
  if degree in degree_connectivity_dict:
    degree_connectivity_dict[degree].append(connectivity)
  else:
    degree_connectivity_dict[degree] = [connectivity]

# Compute the average degree connectivity for each degree value
average_degree_connectivity = {}
for degree, connectivity_values in degree_connectivity_dict.items():
  average_degree_connectivity[degree] = sum(connectivity_values) / len(connectivity_values)

# Print the average degree connectivity for each degree
for degree,connectivity in average_degree_connectivity.items():
  print(f"Degree {degree}: Average Degree Connectivity = {connectivity}")

```

##Network Construction

**spring_layout**:

Position nodes using Fruchterman-Reingold force-directed algorithm.

The algorithm simulates a force-directed representation of the network treating edges as springs

holding nodes close, while treating nodes as repelling objects, sometimes called an

 anti-gravity force. Simulation continues until the positions are close to an equilibrium.

```python
G = nx.Graph()

G.add_edge("a","b",weight=0.6)
G.add_edge("a","c",weight=0.2)
G.add_edge("c","d",weight=0.1)
G.add_edge("c","e",weight=0.7)
G.add_edge("c","f",weight=0.9)
G.add_edge("a","d",weight=0.3)

large_edges = [(u,v) for (u,v,d) in G.edges(data=True) if d["weight"] >= 0.5]
small_edges = [(u,v) for (u,v,d) in G.edges(data=True) if d["weight"] < 0.5]

pos = nx.spring_layout(G,seed= 7) # positions for all nodes - seed for reproducibility
# draw nodes
nx.draw_networkx_nodes(G,pos,node_size=650)
# draw edges
nx.draw_networkx_edges(G,pos,edgelist=large_edges,width=6)
nx.draw_networkx_edges(G,pos,edgelist=small_edges,width=6,alpha=0.5,
                       edge_color='blue',style='dashed')

# label nodes
nx.draw_networkx_labels(G,pos,font_size=20,font_family='sans-serif')

#label edges
edge_labels = nx.get_edge_attributes(G,"weight")
nx.draw_networkx_edge_labels(G,pos,edge_labels)

ax = plt.gca()
ax.margins(0.08)
plt.axis("off")
plt.tight_layout()
plt.show()
```
